```python
class Persona:
    def __init__(self, nombre, edad, dni):
        self.nombre = nombre
        self.edad = int(edad)
        self.dni = dni

    @property
    def nombre(self):
        return self._nombre
    
    @nombre.setter
    def nombre(self, nuevo_nombre):
        if isinstance(nuevo_nombre, str) and len(nuevo_nombre) > 0:
            self._nombre = nuevo_nombre
        else:
            raise ValueError("El nombre debe ser una cadena no vacía.")

    @property
    def edad(self):
        return self._edad
    
    @edad.setter
    def edad(self, nueva_edad):
        if isinstance(nueva_edad, int) and 0 <= nueva_edad <= 120:
            self._edad = nueva_edad
        else:
            raise ValueError("La edad debe ser un número entero entre 0 y 120.")

    @property
    def dni(self):
        return self._dni
    
    @dni.setter
    def dni(self, nuevo_dni):
        if isinstance(nuevo_dni, str) and len(nuevo_dni) == 9 and nuevo_dni[:-1].isdigit() and nuevo_dni[-1].isalpha():
            self._dni = nuevo_dni
        else:
            raise ValueError("El DNI debe tener 8 dígitos seguidos de una letra.")

    def mostrar(self):
        print(f"El nombre es {self.nombre}, la edad es {self.edad} y su dni es el de {self.dni}")

    def es_mayor_de_edad(self):
        return f"{self.nombre} es mayor de edad " if self.edad >= 18 else f"{self.nombre} es menor de edad"
        
p1 = Persona("Raul", 20, "43589358N")
p1.mostrar()
print(p1.es_mayor_de_edad())
```

```python
class Cuenta(Persona):
    def __init__(self, nombre, cantidad=0.0):
        self.titular = nombre
        self.saldo = float(cantidad)
        
    @property
    def titular(self):
        return self._titular
        
    @titular.setter
    def titular(self, nuevo_titular):
        self._titular = nuevo_titular if isinstance(nuevo_titular, str) and len(nuevo_titular) > 0 else ValueError("El Titular debe ser una cadena no vacía.")
            
    @property
    def saldo(self):
        return self._saldo
    
    @saldo.setter
    def saldo(self, nuevo_saldo):
        self._saldo = nuevo_saldo
        
    def mostar(self):
        print(f" Tienes {self.saldo}€ a tu cuenta.")
    
    def ingresar(self, cantidad):
        self.saldo += cantidad; print("Dinero ingresado") if cantidad > 0.1 else print("El numero tiene que ser superior a 0.1")
            
    def retirada(self, cantidad):
        
         if cantidad > self.saldo: self.saldo -= cantidad; print(f"Estas en numeros rojos: {self.saldo}")
         else: print(f"Tu saldo es de {self.saldo}");self.saldo -= cantidad
    
p2 = Persona('Juan', 39, "54863254P")

cuenta1 = Cuenta(p1,500)
cuenta2 = Cuenta(p2,0.1)

print("Salarios")
cuenta1.mostar()
cuenta2.mostar()

print('Ingresos')
cuenta1.ingresar(0)
cuenta2.ingresar(645.3)
print("Salarios2")
cuenta1.mostar()
cuenta2.mostar()

print('Retiradas')
cuenta1.retirada(508.5)
cuenta2.retirada(50.7)

            
    
```

```python
class CuentaJoven(Cuenta, Persona):
    def __init__(self, nombre, edad, saldo, bonificacion=0, dni=""):
        Persona.__init__(self, nombre, edad, dni)
        Cuenta.__init__(self, nombre, saldo)
        self.bonificacion = bonificacion
    
    @property
    def bonificacion(self):
        return self._bonificacion
    
    @bonificacion.setter
    def bonificacion(self, nueva_bonificacion):
        if 0 <= nueva_bonificacion <= 100:
            self._bonificacion = nueva_bonificacion
        else:
            raise ValueError('La bonificación debe estar entre 0 y 100')

    def esTitularValido(self):
        return 18 <= self.edad < 25
    
    def retirada(self, cantidad):
        if self.esTitularValido():
            super().retirada(cantidad)
        else:
            print("El titular no es válido para retirar dinero en una Cuenta Joven")

    def mostrar(self):
        return f"Cuenta Joven de {self.nombre}, saldo disponible: {self.saldo}€, bonificación: {self.bonificacion}%"

p1 = Persona("Juan", 20, "12345678A")
cuenta_joven_p1 = CuentaJoven(p1.nombre, p1.edad, 1000, 10, p1.dni)

print(cuenta_joven_p1.mostrar())

cuenta_joven_p1.retirada(100)

p2 = Persona("Ana", 30, "87654321B")
cuenta_joven_p2 = CuentaJoven(p2.nombre, p2.edad, 2000, 15, p2.dni)
print(cuenta_joven_p2.mostrar())
cuenta_joven_p2.retirada(200)
```
